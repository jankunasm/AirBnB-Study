```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../files/AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

air_bnb['neighbourhood_group'].describe()

# Answer: 5 unique neighborhood groups and Manhattan shows up the most.
```

```python
# Are private rooms the most popular in manhattan?

air_bnb.groupby(['room_type', 'neighbourhood_group']).count().sort_values('reviews_per_month', ascending = False)

# Answer: Private rooms seem to be more popular in Brooklyn based on 
# number of listings and reviews.                                                                   
                                                                
```

```python
# Which hosts are the busiest based on their reviews?

air_bnb.sort_values('number_of_reviews', ascending = False)

# Answer: Dona with 629 reviews
```

```python
#Which neighorhood group has the highest average price?

air_bnb.groupby('neighbourhood_group').describe()[['price']]

# Answer: Manhattan has the highest average price.
```

```python
# Which neighbor hood group has the highest total price?

air_bnb.groupby(['neighbourhood']).sum()[['price']].sort_values('price', ascending = False)

# Answer: Williamsburg has the highest total price.
```

```python
#Which top 5 hosts have the highest total price?

air_bnb.groupby(['host_name', 'price']).head().sort_values('price', ascending = False)

# Answer: Jelenna, Erin, Katherine, Matt, and Olson
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

grouped = air_bnb[air_bnb['number_of_reviews'] >= 100].groupby(['host_name', 'availability_365']).head().sort_values('availability_365', ascending = True)
grouped[grouped['availability_365'] == 0]

# Answer: Currently 162 hosts have zero availability with > 100 reviews
```

```python
# What host has the highest total of prices and where are they located?

air_bnb.groupby(['host_name']).sum()[['price']].sort_values('price', ascending = False)

# Answer: Sonder (NYC) Manhattan
```

```python
# When did Danielle from Queens last receive a review?

# grouped = air_bnb[air_bnb['host_name'] == 'Danielle']
# grouped[grouped['neighbourhood_group'] == 'Queens']
# Host Id 26432133

air_bnb[air_bnb['host_id'] == 26432133]

# Answer: last Review 2019-07-08
```
